```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy import stats
sns.set(color_codes=True)
sns.set(style="ticks")
```

```python
df = pd.read_csv("cmc.csv")
df.columns = ["w_age", "w_ed", "h_ed", "num_kid", "w_islam", "home", "h_job",
              "sol", "med_ex", "cont"]
print(df.head)
```

**<b>Choose one variable and plot that variable four different ways.</b>

```python
sns.relplot(x="w_age", y="num_kid", data=df);
```

This Seaborn scatterplot does a nice job of giving a general picture of where some values may or may not fall.  It is easy to see that many more women have fewer children than those who have many children.  However, one cannot see how many of each value are in the dataset.  All you really know is that where there is no dot, there are no values.  If there is a dot present, it could mean there is one value or it could mean that there are a million.

```python
sns.relplot(x="w_age", y="num_kid", kind="line", data=df);
```

This Seaborn lineplot shows the average ranges of women's ages and number of children, and the shaded areas represent the confidence intervals for each value.  This visualization is good for some general trends, but difficult to extrapolate much valuable data with such a large dataset (with the exception of whether the values are highly varied or fairly similar).

```python
sns.jointplot(x="w_age", y="num_kid", data=df);
```

This Seaborn jointplot is a scatterplot with marginal histograms, which makes this scatterplot much more valuable than the first one.  Now one can see how many values are in each row/column, which is exactly the weakness of the previous plot.  I would think that this is a much better option in almost every circumstance than just a simple relplot.

```python
sns.jointplot(x="w_age", y="num_kid", data=df, kind="hex");
```

This Seaborn jointplot takes the scatterplot data and places them in hexagonal bins that are shaded based on the number of values within each hexagon.  This allows one to see the structure of the dataset with an added level of clarity.

**<b>Choose two continuous variables, and plot them three different ways.</b>

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

This Seaborn distribution plot shows a combination of a histogram (as a bar graph) and a kernal density estimate, or kde (as a line graph).  This plot gives two dimensions of the dataset, which makes it better than each individual plot.

```python
sns.distplot(x, bins=35, kde=False, rug=True);
```

This Seaborn distribution plot show a combination of a histogram (this time with more bins than the previous graph) and a kde plot in the form of a rugplot.  Instead of the line graph in the previous graph, the kde data is represented by small lines of varying thickness/darkness, representing greater values.

```python
sns.kdeplot(x, shade=True);
```

This Seaborn plot is simply the kdeplot from the first graph in this section.  As stated in that answer, it is not nearly as useful as the combined histogram, but if all you need is the kdeplot, then it is perfect!

**<b>Choose one continuous variable and one categorical variable, and plot them six different ways.</b>

```python
sns.jointplot(x="w_age", y="cont", data=df);
```

In this Seaborn jointplot, the categorical variable is listed as "cont" and has a value of either 1, 2, or 3.  The continuous variable is listed as "w_age" and has values ranging within 15 and 50.  This is better than a singular histogram because one can see the values in two dimensions (how many women in the dataset are over 45 vs. how many women identified within category 2).  The scatterplot is not nearly as useful in this example, due to the categorical nature (again, the only thing we see for certain is where values are not in each data set, not any information about how many values are at each datapoint.

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.w_age, df.cont, cmap=cmap, n_levels=60, shade=True);
```

This Seaborn kde plot has a very interesting appearance, showing the same data as the previous graph in a more fluid medium.  One can see that there are many women within the ages of 20 to 30, and most of them are either in the 1 or 3 cont category.  

```python
g = sns.jointplot(x="w_age", y="cont", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("Wife's Age", "Contraception");
```

This Seaborn kde plot shows the line graph on the axes of the graph and a scatterplot (shown as white '+' marks) with a graph aligning the marginal axes with the joint axes.  Again, adding additional ways to show the data makes this an informative, albeit somewhat complicated. 

```python
sns.lmplot(x="w_age", y="cont", hue="cont", data=df, palette="Set1");
```

This Seaborn linear regression model shows the same data with colors added to the categorical data.  This is not especially useful in this particular case, although it could be very effective with different parameters.  

```python
sns.catplot(x="w_age", y="cont", hue="w_ed",
            kind="violin", data=df);
```

This Seaborn violin plot is showing the same data with an added variable of 'w_ed' that is color coded, but due to the number of parameters on the x-axis, it is too difficult to read.  With fewer values, this could be a very effective visualization.

```python
sns.catplot(x="w_age", y="cont", hue="w_ed", kind="bar", data=df);
```

This cat plot is similar to the violin plot, showing the same data, but having the colored portions extend down to the x-axis.  Similarly, the number of values makes this too difficult to read.
